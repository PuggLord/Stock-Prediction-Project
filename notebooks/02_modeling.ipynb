# Modeling Result Explorer
Loads `reports/results.json` and summarizes model and backtest outputs.

```python
from pathlib import Path
import json
import pandas as pd

ROOT = Path.cwd()
if not (ROOT / 'reports').exists():
    ROOT = ROOT.parent

results_path = ROOT / 'reports' / 'results.json'
results = json.loads(results_path.read_text(encoding='utf-8'))
results.keys()
```

```python
classification = results['classification']
classification_table = pd.DataFrame([
    {'model': 'Logistic Regression', **classification['logistic_regression']},
    {'model': 'Random Forest', **classification['random_forest']}
])
classification_table
```

```python
backtest = results['backtest']
backtest_table = pd.DataFrame([
    {'strategy': 'Logistic Regression', **backtest['logistic_regression']},
    {'strategy': 'Random Forest', **backtest['random_forest']},
    {'strategy': 'Buy and Hold', **backtest['buy_and_hold']}
])
backtest_table
```
